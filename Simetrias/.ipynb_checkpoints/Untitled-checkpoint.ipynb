```python
p=o3d.io.read_point_cloud('D:\Documentos\INNOVATE\lib\symmetry_detection_python\Linea 12.pts')
dp=p
dp.estimate_normals(search_param=o3d.geometry.KDTreeSearchParamHybrid(radius=0.1, max_nn=30))
o3d.visualization.draw_geometries([dp],point_show_normal=False)

def display_inlier_outlier(cloud, ind):
    inlier_cloud = cloud.select_by_index(ind)
    outlier_cloud = cloud.select_by_index(ind, invert=True)

    print("Showing outliers (red) and inliers (gray): ")
    outlier_cloud.paint_uniform_color([1, 0, 0])
    inlier_cloud.paint_uniform_color([0.8, 0.8, 0.8])
    o3d.visualization.draw_geometries([inlier_cloud, outlier_cloud])
cl, ind = dp.remove_statistical_outlier(nb_neighbors=5,std_ratio=1.5)
display_inlier_outlier(dp, ind)
dp=cl
dp=dp.voxel_down_sample(voxel_size=0.015)
dp.estimate_normals(search_param=o3d.geometry.KDTreeSearchParamHybrid(radius=0.1, max_nn=30))
o3d.visualization.draw_geometries([dp])
print(np.asarray(dp.points).shape)
```

```python
#Generate Perfect form
R=4
L=np.arange(10000)/1000
r=np.ones(10000)*4
th=(2*np.pi/10000)*np.arange(10000)
z=L
x=r*np.sin(th)
y=r*np.cos(th)
form=np.concatenate((x.reshape(-1,1),y.reshape(-1,1),z.reshape(-1,1)),axis=1)
trans_g=o3d.geometry.PointCloud()
trans_g.points = o3d.utility.Vector3dVector(form)
o3d.visualization.draw_geometries([trans_g])
```
